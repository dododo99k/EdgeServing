```python
import pickle
import time

# read ./logs/lam152_20/multi_model_diag_ours.pkl
with open("./logs/lam152_180/multi_model_diag_ours_lam180.pkl", "rb") as f:
    data = pickle.load(f)   
```

```python
batch_diag = data["batch_diag"]
resnet50_list = []
resnet101_list = []
resnet152_list = []

for dict in batch_diag:
    if dict["model_name"] == "ResNet50":
        resnet50_list.append(dict)
    elif dict["model_name"] == "ResNet101":
        resnet101_list.append(dict)
    elif dict["model_name"] == "ResNet152":
        resnet152_list.append(dict)
```

```python
# get the different of actual_infer_ms and pred_infer_ms
differences = [[], [], []]  # resnet50, resnet101, resnet152
for i, lst in enumerate([resnet50_list, resnet101_list, resnet152_list]):
    for dict in lst:
        diff = dict["actual_infer_ms"] - dict["pred_infer_ms"]
        differences[i].append(diff)
```

```python
for li in differences:
    print(f"Number of samples: {len(li)}")
    print(f"Mean difference: {sum(li)/len(li):.2f} ms")
    print(f"Max difference: {max(li):.2f} ms")
    print(f"Min difference: {min(li):.2f} ms")
    print(f"Variance: {sum((x - sum(li)/len(li))**2 for x in li)/len(li):.2f} ms^2")
```

```python
# find index of max difference in resnet50
max_diff_index = differences[0].index(max(differences[0]))
print("Max difference index in resnet50:", max_diff_index)
print("ResNet50 batch diag at max difference index:")
# print dict in many lines
for key, value in resnet50_list[max_diff_index].items():
    print(f"{key}: {value}")
```

```python
# find index of max difference in resnet50
max_diff_index = differences[1].index(max(differences[1]))
print("Max difference index in resnet101:", max_diff_index)
print("ResNet101 batch diag at max difference index:")
# print dict in many lines
for key, value in resnet101_list[max_diff_index].items():
    print(f"{key}: {value}")
```
